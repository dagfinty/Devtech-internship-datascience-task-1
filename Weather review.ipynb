```python
import pandas as pd

df = pd.read_csv("weather_data.csv")
df
```

```python
avg_temp = df['Temperature'].mean()
print(f"Average Temperature: {avg_temp:.2f}°C")


hottest_day = df.loc[df['Temperature'].idxmax()]
print(f"Hottest Day: {hottest_day['Date']} ({hottest_day['Temperature']}°C)")


coldest_day = df.loc[df['Temperature'].idxmin()]
print(f"Coldest Day: {coldest_day['Date']} ({coldest_day['Temperature']}°C)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nicer
sns.set(style="whitegrid")

# Create a line plot
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Temperature', data=df, marker='o')
plt.title("Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)  # Rotate dates for readability
sns.lineplot(x='Date', y='Humidity', data=df, marker='o')
plt.title("Humidity and Temprture Over Time")
plt.xlabel("Date")
plt.ylabel("Humidity (°C)")
plt.xticks(rotation=46)  # Rotate dates for readability
plt.show()
```

```python
print("Weather Analysis Summary")
print("------------------------")
print(f"Average Temperature: {avg_temp:.2f}°C")
print(f"Hottest Day: {hottest_day['Date']} ({hottest_day['Temperature']}°C)")
print(f"Coldest Day: {coldest_day['Date']} ({coldest_day['Temperature']}°C)")

```
